```python
#path of the file to read
flight_filepath=r"C:\Users\Praku\Week 13 Statistics\data-for-datavis\flight_delays.csv"

# read the file into a variable sportify_data

flight_data=pd.read_csv(spotify_filepath,index_col="Month")
```

```python
flight_data.head(10)
```

```python
flight_data.tail(10)
```

```python
plt.figure(figsize=(10,6))

plt.title("Average arrival delay for spirit airlines flight")

sns.barplot(x=flight_data.index,y=flight_data['NK'])

plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(14,7))

plt.title("Average arrival Delay for Each Ariline , by month")

sns.heatmap(data=flight_data,annot=True)

plt.xlabel("Airline")
```

```python
plt.figure(figsize=(14,7))

corr=flight_data.corr()

ax= sns.heatmap(corr,vmin=-1,vmax=1,center=0,annot=True)
```
